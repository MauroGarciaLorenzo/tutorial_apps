```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def plot_a_line (a,b,c):
    x = np.linspace(-4, 4, 100)
    y = a * x ** 2 + b * x + c  # Compute y for all x values
    plt.plot(x, y)
```

```python
def plot_another_line():
    x = np.linspace(-3, 3, 100)
    y = (x + 2) * (x - 1) * (x - 2)
    plt.plot(x, y, 'r--')
    plt.plot([-2, 1, 2], [0, 0, 0], 'bo', markersize=10)
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.title('Another Python line');
```

```python
plot_a_line(1,1, -6)
```

```python
plot_another_line()
```
